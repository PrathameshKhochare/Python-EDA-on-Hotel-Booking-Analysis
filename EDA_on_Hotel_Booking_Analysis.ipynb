## 3. ***Data Wrangling***

##**Data Cleaning**

````python
# Create a copy of the current dataset and assigning to df1
df1=df.copy()
````

````python
# sorting null values in descending manner
df1.isna().sum().sort_values(ascending=False)[:6].reset_index().rename(columns={'index':'column',0:'Null values'})
````

````python
# from above operations it is found that column country,children,agent and company have null values
# As Dtype of children,company and agent is float, we can replace null values with 0. Also Dtype of country is object we can replace null values with 'other'
df1['company'].fillna(0,inplace=True)
df1['agent'].fillna(0,inplace=True)
df1['children'].fillna(0,inplace=True)
df1['country'].fillna('Other',inplace=True)
````

````python
#null check
df1.isna().sum()
````

````python
# Checking for duplicate rows
df1.duplicated().value_counts()
````

````python
# above operation shows datasheet has 31994 duplicated rows, we can drop this from df1
df1 = df1.drop_duplicates()
````

````python
# Duplicate check
print(df1.duplicated().value_counts())
#shape
df1.shape
````

### **Feature Engineering**

We can merge Stay in week days and weekend as total Stay. also we can merge number of childern,adults and babies as total people.

We can add revenue column which is miltiplication of adr and total stay.

````python
# total_stay = weekdays + weekends
df1['total_stay'] = df1['stays_in_week_nights'] + df1['stays_in_weekend_nights']
#total_people = adults + children + babies
df1['total_people'] = df1['children'] + df1['adults'] + df1['babies']
# Adding Revenue, Revenue = adr*total_stay
df1['revenue'] = df1['adr'] * df1['total_stay']
````

Adding new column which shows room type allocated is same as assigned or not

````python
# new column
df1['is_reserved_room_assigned'] = df1.apply(lambda x: 'same room type assigned' if x['reserved_room_type'] == x['assigned_room_type'] 
                                  else 'different room type assigned' , axis =1)
````

````python
# varification of updation
print(list(df1.columns))
df1.shape
````

There are some rows in total_people are zero. so we can drop this rows.

````python
# rows of total people with value zero
df1[df1['total_people'] == 0].shape
````

````python
# so we can drop this 166 rows
df1.drop(df1[df1['total_people'] == 0].index,inplace=True)
````

````python
# check of drop operation
df1[df1['total_people'] == 0].shape
````

Converting Datatypes for some columns approprietly

As columns 'children','agent','company' and 'total_people'and 'revenue' have datatype float instead of int we can convert float to int. Also column 'reservation_status_date' has datatype object, we can convert into date type


````python
# float to int
df1[['agent','company','children','total_people','revenue','adr']] = df1[['agent','company','children','total_people','revenue','adr']].astype('int64')

#Object to Date type
df1['reservation_status_date'] = pd.to_datetime(df1['reservation_status_date'], format = '%Y-%m-%d')
````


As columns is_cancel and is_repeated_guests have boolean values, we can convert them to string values for better understanding

````python
#boolen to string
df1['is_canceled'] = df1['is_canceled'].replace([0,1],['not canceled','cancled'])
df1['is_repeated_guest'] = df1['is_repeated_guest'].replace([0,1],['not repeated guest','repeated guest'])
````

````python
# verification
df1.info()
````

````python
# shape 
df1.shape
````
